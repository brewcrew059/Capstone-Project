```python
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues)
nearby_venues.head()
```

```python
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]
nearby_venues.head()
```

```python
name_list = []
for i in nearby_venues['venue.categories']:
    name_list.append(i[0]["name"])
    
name_list
```

```python
nearby_venues['venue.categories'] = name_list
nearby_venues.head()
```

```python
# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues.head()
```

```python
#Filter out non-food related venues
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Accessories Store'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Bowling Alley'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Business Service'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Clothing Store'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Gift Shop'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Gym / Fitness Center'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Hotel'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Park'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Pharmacy'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Rental Car Location'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Shipping Store'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Theater'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Thrift / Vintage Store'].index, inplace = True)
nearby_venues.drop(nearby_venues[nearby_venues['categories'] == 'Train Station'].index, inplace = True)
```

```python
nearby_venues.reset_index(inplace = True)
nearby_venues.drop('index', axis= 1, inplace = True)
nearby_venues
```

```python
nearby_venues.groupby('categories').count()
```

```python
whitefish_onehot = pd.get_dummies(nearby_venues[['categories']], prefix="", prefix_sep="")

whitefish_onehot.reset_index()

whitefish_onehot
```

```python
whitefish_grouped = whitefish_onehot.mean().reset_index()
whitefish_grouped.rename(columns={0: "freq"}, inplace=True)
whitefish_grouped
```

```python
whitefish_grouped.sort_values(by='freq', inplace=True)
whitefish_grouped
```
